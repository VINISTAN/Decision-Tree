```python
# Import library files for access .csv files
import pandas as pd
```

```python
# Read Dataset using pandas
dataset = pd.read_csv("50_Startups.csv")
```

```python
#visualize the dataset
dataset
```

```python
# Convert categorical data into neumerical format using get dummies method
dataset = pd.get_dummies(dataset,drop_first = True)
```

```python
# Visualize the updated dataset
dataset
```

```python
# get the available columns from dataset
dataset.columns
```

```python
# Split input (Independent) and output (Dependent) data
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#Visualize the input (Independent) data
independent
```

```python
# Dependent Variable
dependent = dataset[['Profit']]
```

```python
# Visualize the dependent variable
dependent
```

```python
# Split train dataset and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30,random_state = 0)
```

```python
#visualize trin and test dataset
x_train,x_test,y_train,y_test
```

```python
#  Model creation using Decision Tree 
from sklearn.tree import DecisionTreeRegressor
```

```python
#Assign DecisionTreeRegressor class to variable
regressor = DecisionTreeRegressor()
```

```python
# Model Creation
regressor = regressor.fit(x_train,y_train)
```

```python
#Visualize the decision tree structure
#import ploting libraries using matplotlib and sklearn
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
# Visualize the tree structure after the model creation
tree.plot_tree(regressor)
plt.show()
```

```python
# Predict the output variable using input - 30% test data
y_predict = regressor.predict(x_test)
y_predict
```

```python
# Evaluation using R^2 value
# Import Library files 
from sklearn.metrics import r2_score
```

```python
# Test the r2_score value
r2_score = r2_score(y_test,y_predict)
```

```python
# Visualize the r2_score
r2_score
```

```python
# Save the model
import pickle
```

```python
file_name = "Decision Tree startup prediction Model.sav"
pickle.dump(regressor,open(file_name,"wb"))
```
